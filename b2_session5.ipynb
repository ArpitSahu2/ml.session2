```python
#groups and counts-gdp population and life expentancy
import pandas as pd

# Step 1: Load your dataset
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv")

# Step 2: Clean the dataset
# Example: drop missing values and duplicates
df_clean = df.dropna().drop_duplicates()

# Step 3: Now you can use df_clean
print(df_clean["continent"].value_counts())

continent_summary = df_clean.groupby("continent")[[
    "lifeExp",
    "gdpPercap",
    "pop"
]].mean()

print(continent_summary)
```

```python
#using the lib matplotlib i
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Group by continent and calculate average life expectancy
avg_life_exp = df.groupby("continent")["lifeExp"].mean()

# Plot bar chart
plt.figure(figsize=(8,6))
avg_life_exp.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Average Life Expectancy by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=45)
plt.show()
```

```python
#temporal?
#line chart life expectancy over time for 3 countries india,china,unitedstates
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Filter for the three countries
countries = ["India", "China", "United States"]
df_filtered = df[df['country'].isin(countries)]

# Plot line chart
plt.figure(figsize=(10,6))
for country in countries:
    country_data = df_filtered[df_filtered['country'] == country]
    plt.plot(country_data['year'], country_data['lifeExp'], marker='o', label=country)

# Chart formatting
plt.title("Life Expectancy Over Time (India, China, United States)")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#scatter plot gdp per capita vs life expectancy
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Scatter plot: GDP per capita vs Life Expectancy
plt.figure(figsize=(9,5))
plt.scatter(df['gdpPercap'], df['lifeExp'], alpha=0.6, c='grey')

plt.title("GDP per Capita vs Life Expectancy")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.grid(True)
plt.show()

```
